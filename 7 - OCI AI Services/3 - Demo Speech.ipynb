# Lesson: Using OCI Speech in the Console – Creating a Transcription Job

## Introduction
Welcome to this lesson on **using OCI Speech in the Oracle Cloud Infrastructure (OCI) Console**.  
In this session, we’ll walk through the steps required to **create and run a speech-to-text transcription job** using the OCI Speech service.  

You’ll learn how to:
- Navigate to the Speech service in the OCI Console  
- Prepare data in Object Storage  
- Create and execute a transcription job  
- View and interpret the transcription results  

---

## Step 1: Accessing the OCI Speech Service
Start by signing in to the **OCI Console**.

1. From the main console page, open the **Navigation Menu**.  
2. Go to **Analytics & AI**.  
3. Under **AI Services**, select **Speech**.  

This will take you to the **OCI Speech service console page**, where you can manage transcription jobs, review previous tasks, and create new ones.

---

## Step 2: Preparing Data in Object Storage
Before creating a transcription job, you need to have your **audio files** stored in **Object Storage**.

- OCI Speech processes data directly from your Object Storage bucket.  
- Supported audio formats include **.wav**, **.mp3**, and **.mp4** (depending on encoding).  
- Ensure that your bucket and files are in the **same compartment** that you plan to use for the transcription job.

In this example, the instructor has already uploaded a `.wav` file to an Object Storage bucket within the selected compartment.

---

## Step 3: Creating a Transcription Job
Now, let’s create the transcription job.

1. On the **Speech service console page**, click **Create Transcription Job**.  
2. Enter a name for the job — here we’ll call it **Training**.  
3. Confirm that the correct **compartment** is selected.  
4. Choose the **Object Storage bucket** that contains your audio file.  

The console will display the available files in the selected bucket.  
Select the desired **.wav** file (in this example, the instructor’s previously uploaded file).

---

## Step 4: Running the Job
Once the configuration is complete:

- Click **Run Job** to start the transcription process.  
- The job typically takes **only a few seconds** to complete (depending on file size).  
- You can monitor the job status on the console — it will move from **In Progress** to **Succeeded** once complete.

Because the instructor had run this job several times before, the result appeared very quickly.

---

## Step 5: Viewing the Transcription Results
After the job status changes to **Succeeded**, select the file to view the transcription output.

The resulting text from the example is:

> “I’m having an issue with my wireless headphones. I think my Bluetooth connectivity to the TV is not working. I have high-quality headphones. I know the issue is not with them.”

And then a second voice responds:

> “Hi, thanks for contacting support. I’m sorry to hear about your Bluetooth issue.”

This demonstrates how **OCI Speech** accurately transcribes multiple speakers and maintains the **flow and clarity** of the conversation.

---

## Step 6: Observing Additional Features
When reviewing the transcription results, note a few important enhancements:

- **Punctuation:**  
  The output includes proper punctuation marks (periods, commas, and capitalization), making the text easier to read and interpret.

- **Normalization:**  
  Numbers and special terms are automatically formatted.  
  For example, instead of spelling out “one hundred percent,” the transcription shows **100%**, reflecting natural written conventions.

These features greatly improve the readability and usability of the transcribed text for downstream processing.

---

## Summary
In this lesson, you learned how to use **OCI Speech via the Console** to create and run a transcription job.  

We covered how to:
- Access OCI Speech from the **Analytics & AI** section  
- Prepare and select audio data from **Object Storage**  
- Create, run, and monitor a **transcription job**  
- Review accurate, punctuated, and normalized text output  

By following these steps, you can easily transform audio recordings into high-quality text — ready for analysis, captioning, or further AI processing.
